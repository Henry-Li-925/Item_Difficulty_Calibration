```python
# Problem Dimension (Non-NA)
problem_dim = problem_df.drop_duplicates(subset=[PROBLEM_ID, ASSISTMENT_ID, SKILL_CODE])
dim_count = problem_dim.groupby(PROBLEM_ID)[SKILL_CODE].count()
multi_dim_prob = dim_count[dim_count > 1].reset_index().rename(columns={SKILL_CODE: 'dim_count'})
print('Number of problems with multiple dimensions (Non-NA): ', len(multi_dim_prob))
```

```python
# Problem Dimension
na_problem_dim = problem_dim.copy() 
na_problem_dim[SKILL_CODE] = na_problem_dim[SKILL_CODE].mask(na_problem_dim[SKILL_CODE].isna(), 'UNK')
na_dim_count = na_problem_dim.groupby(PROBLEM_ID)[SKILL_CODE].count()
na_multi_dim_prob = na_dim_count[na_dim_count > 1].reset_index().rename(columns={SKILL_CODE: 'dim_count'})
print('Number of problems with multiple dimensions (Count NA): ', len(na_multi_dim_prob))
```

```python
uni_problem = problem_dim[~problem_dim[PROBLEM_ID].isin(na_multi_dim_prob[PROBLEM_ID])]
print('Number of unidimensional problems (Non-NA): ', len(uni_problem[uni_problem[SKILL_CODE].notna()]))
print('Number of unidimensional problems (NA): ', len(uni_problem[uni_problem[SKILL_CODE].isna()]))
```

#### Problem Response

```python
problem_resp_count['response_count'].describe()
```

```python
# counts number of problems that have response_count = 1
ans_1 = len(problem_resp_count[problem_resp_count['response_count'] == 1])
# counts number of problems that have response_count between 2-10
ans_10 = len(problem_resp_count[(problem_resp_count['response_count'] > 2) & (problem_resp_count['response_count'] < 10)])
# counts number of problems that have response_count between 10-50
ans_50 = len(problem_resp_count[(problem_resp_count['response_count'] >= 10) & (problem_resp_count['response_count'] < 50)])
# counts number of problems that have response_count between 50-100
ans_100 = len(problem_resp_count[(problem_resp_count['response_count'] >= 50) & (problem_resp_count['response_count'] < 100)])
# counts number of problems that have response_count between 100-200
ans_200 = len(problem_resp_count[(problem_resp_count['response_count'] >= 100) & (problem_resp_count['response_count'] < 200)])
# counts number of problems that have response_count between 200-500
ans_500 = len(problem_resp_count[(problem_resp_count['response_count'] >= 200) & (problem_resp_count['response_count'] < 500)])
# counts number of problems that have response_count between 500-1000
ans_1000 = len(problem_resp_count[(problem_resp_count['response_count'] >= 500) & (problem_resp_count['response_count'] < 1000)])
# counts number of problems that have response_count more than 1000
ans_1000_plus = len(problem_resp_count[problem_resp_count['response_count'] >= 1000])

ans_count = pd.Series([ans_1, ans_10, ans_50, ans_100, ans_200, ans_500, ans_1000, ans_1000_plus], index=['1', '2-10', '10-50', '50-100', '100-200', '200-500', '500-1000', '1000+'])
# Convert Series to DataFrame for seaborn plotting
ans_count_df = pd.DataFrame({'count': ans_count.values, 'range': ans_count.index})
ans_count_df
```

```python
sns.barplot(data=ans_count_df, x='range', y='count')
# label each bar with the count
for index, row in ans_count_df.iterrows():
    plt.text(row['range'], row['count'], row['count'], ha='center', va='bottom')
plt.show()
```
